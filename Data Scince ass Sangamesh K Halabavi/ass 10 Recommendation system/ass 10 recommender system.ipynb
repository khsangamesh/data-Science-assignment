```python
user_books = books.pivot_table(index='UserID',
                        columns='BookTitle',
                         values='BookRating')
```

```python
user_books.fillna(0,inplace=True)
```

```python
user_books
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_number=1-pairwise_distances(user_books.values,metric='cosine')
```

```python
user_number
```

```python
user_number_df=pd.DataFrame(user_number)
```

```python
user_number_df
```

```python
user_number_df.index=books.UserID.unique()
user_number_df.columns=books.UserID.unique()
```
